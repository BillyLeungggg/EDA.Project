```python
sleep_info["Age Group"] = sleep_info["Age"].apply(lambda x: "7-11" if x < 12 else "12-18" if x < 19 else "19-29" if x<30 else "30-45" if x < 46 else "46-64" if x<65 else "More than Age 64")
```

```python
sleep_info["Sleeping Sufficiency"] = sleep_info["Sleep duration"].apply(lambda x: "Enough" if x >= 7 else "Not enough")
```

```python
sleep_info.head(1)
```

```python
sleep_info["Sleeping Quality"] = sleep_info["Deep sleep percentage"].apply(lambda x: "Good" if x >= 25 else "Bad")
```

```python
# Create a bar chart using Seaborn
sns.set(style="whitegrid")  # Set the style of the plot
plt.figure(figsize=(8, 6))  # Set the size of the plot

# Create the bar chart
#sns.barplot(x='Sleeping Sufficiency', y=a, palette="rocket")
sns.countplot(x="Sleeping Sufficiency",data=sleep_info,palette="pastel",edgecolor = "black")

# Set labels and title
plt.xlabel("Sleeping Sufficiency")
plt.ylabel("People")
plt.title("People' Sleeping Efficiency")

# Show the chart
plt.show()
```

```python
#sleep_info.groupby('Age Group')['Sleeping Quality'].value_counts().plot.pie()
print(sleep_info['Awakenings'].value_counts().sort_index())
sizes = sleep_info['Awakenings'].value_counts().sort_index()
#sleep_info['Awakenings'].value_counts()
colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#FF00FF', '#00FFFF']
labels = ['0', '1', '1.64', '2','3','4']
# Create a pie chart with Seaborn
plt.figure(figsize=(8, 6))

# Specify the labels, sizes, and colors
plt.pie(sizes,  colors=colors, autopct='%1.1f%%', startangle=90)
plt.legend(labels, loc="upper left")
# Set aspect ratio to be equal so that pie is drawn as a circle
plt.axis('equal')

# Set title
plt.title("Frequency of Awakenings")

# Show the pie chart
plt.show()
```

```python
#sleep_info.groupby('Age Group')['Sleeping Quality'].value_counts().plot.pie()
#print(sleep_info['Caffeine consumption'].value_counts() )
print(sleep_info['Caffeine consumption'].value_counts().sort_index() )
sizes = sleep_info['Caffeine consumption'].value_counts().sort_index()
#sleep_info['Awakenings'].value_counts()
colors = ['gold', 'orange', 'blue', 'yellowgreen', 'silver', 'lightskyblue', 'lightcoral'] #'#FFFF00', '#FF00FF', '#00FFFF', '#800080'
labels = ['0 mg', '23.65 mg', '25 mg', '50 mg','75 mg','100 mg','200 mg']
explode = (0, 0, 0, 0, 0, 0.5, 0.7)

# Create a pie chart with Seaborn
plt.figure(figsize=(8, 6))

# Specify the labels, sizes, and colors
plt.pie(sizes,  colors=colors, autopct='%1.1f%%', startangle=180,explode=explode)

#labels=labels,
plt.legend(labels, loc="best")

# Set aspect ratio to be equal so that pie is drawn as a circle
plt.axis('equal')

# Set title
plt.title("Amount of Caffeine consumed in the 24 hours")

# Show the pie chart
plt.show()
```
